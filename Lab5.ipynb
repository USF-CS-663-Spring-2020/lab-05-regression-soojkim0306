# Toluca

```python
import pandas as pd

df = pd.read_csv('toluca.csv')
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x=df['workHours'], y=df['lotSize'], color = 'blue')
plt.xlabel('workHours')
plt.ylabel('lotSize')
```

```python
import seaborn as sns

rl = sns.residplot('workHours', 'lotSize', data = df)
rl.set(xlabel='Lot size', ylabel = 'Residual')
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(df['workHours'].to_frame(), df['lotSize'].to_frame())
# reg.fit(df.iloc[:0], df.iloc[:,1]) not work cuz it's series
lotSize_pred = reg.predict(df['workHours'].to_frame())
```

```python
plt.scatter(x=df['workHours'], y=df['lotSize'], color = 'blue')
plt.plot(df.iloc[:,1].to_frame(), lotSize_pred)
plt.xlabel('workHours')
plt.ylabel('lotSize')
```

```python
#Coefficient
reg.coef_
```

```python
#Intercept
reg.intercept_
```

# Credit

```python
cred_df = pd.read_csv('credit.csv')
cred_df.head()
```

## Income vs Limit

```python
plt.scatter(x=cred_df['Income'],y=cred_df['Limit'])
plt.xlabel('Income')
plt.ylabel('Limit')

```

```python
rl = sns.residplot('Income', 'Limit', data = cred_df)
rl.set(xlabel='Residual', ylabel='Limit')

reg = LinearRegression()
reg.fit(cred_df['Income'].to_frame(), cred_df['Limit'].to_frame())
# reg.fit(df.iloc[:0], df.iloc[:,1]) not work cuz it's series
pred = reg.predict(cred_df['Income'].to_frame())
```

```python
plt.scatter(x=cred_df['Income'],y=cred_df['Limit'])
plt.plot(cred_df['Income'].to_frame(), pred)
plt.xlabel('Income')
plt.ylabel('Limit')
```

```python
print('Income vs Limit')
print('Coefficient: ', reg.coef_)
print('Intercept : ', reg.intercept_)
```

## Limit vs Rating

```python
plt.scatter(x=cred_df['Rating'],y=cred_df['Limit'])
plt.xlabel('Rating')
plt.ylabel('Limit')
```

```python
rl = sns.residplot('Rating', 'Limit', data = cred_df)
rl.set(xlabel='Residual', ylabel='Limit')

reg = LinearRegression()
reg.fit(cred_df['Rating'].to_frame(), cred_df['Limit'].to_frame())
```

```python
rate_pred = reg.predict(cred_df['Rating'].to_frame())
plt.clf()
plt.scatter(x=cred_df['Rating'],y=cred_df['Limit'])
plt.plot(cred_df['Rating'].to_frame(), rate_pred)
plt.xlabel('Rating')
plt.ylabel('Limit')
```

```python
print('Rating vs Limit')
print('Coefficient: ', reg.coef_)
print('Intercept : ', reg.intercept_)
```
